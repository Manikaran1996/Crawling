```python
import requests
from bs4 import BeautifulSoup
from collections import defaultdict
```

```python
url = 'http://districts.nic.in/'
states_ut_india_page = requests.get(url).text
```

```python
states_ut_india_page_soup = BeautifulSoup(states_ut_india_page, 'html.parser')
```

```python
div = states_ut_india_page_soup.findAll('div', attrs={'class':'uppersec'})
```

```python
anchors = []
print(len(div))
for i in range(len(div)):
    anchors = anchors + div[i].findAll('a')
```

```python
states_to_district_link = {}
for a in anchors:
    states_to_district_link[a.text] = a['href']
```

```python
states_to_district_link
```

```python
districts = defaultdict(list)
for state in states_to_district_link:
    districts_page = requests.get(states_to_district_link[state]).text
    soup = BeautifulSoup(districts_page, 'html.parser')
    div_dis = soup.find('div', attrs={'class':'uppersec'})
    anchors = div_dis.findAll('a')
    for a in anchors:
        districts[state].append(a.text)
```

```python
districts
```

```python
len(districts)
```

```python
import pandas as pd
```

```python
districtsDf = pd.DataFrame.from_dict(districts, orient='index')
```
